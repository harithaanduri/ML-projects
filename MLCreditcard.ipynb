Caping outliers

Employed_year

```python
# Here I am using IQR to remove outliers.
# IQR for Employed_year column
Q1 = creditcard_copy['Employed_years'].quantile(0.25)
Q3 = creditcard_copy['Employed_years'].quantile(0.75)
print(Q3, Q1)
IQR = Q3 - Q1
print(IQR)
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)
```

```python
creditcard_copy["Employed_years"]=np.where(creditcard_copy["Employed_years"]>upper_bound,upper_bound,creditcard_copy["Employed_years"])
creditcard_copy["Employed_years"]=np.where(creditcard_copy["Employed_years"]<lower_bound,lower_bound,creditcard_copy["Employed_years"])
```
